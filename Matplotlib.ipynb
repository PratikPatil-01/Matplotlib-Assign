subplot( m , n , p ) divides the current figure into an m -by- n grid and creates axes in the position specified by p . MATLAB numbers subplot positions by row.

```python
a = np.array([0, 1, 2, 3, 4, 5])
b = np.array([0, 100, 200, 300, 400, 500])

c = np.array([0, 1, 2, 3, 4, 5])
d = np.array([50, 20, 40, 20, 60, 70])

e = np.array([0, 1, 2, 3, 4, 5])
f = np.array([10, 20, 30, 40, 50, 60])

g = np.array([0, 1, 2, 3, 4, 5])
h = np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(a, b)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(c, d, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(e, f, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(g, h, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

for ax in axs.flat:
    ax.label_outer()
```

#4

Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data. 

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Barplot")
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.ylabel("Company")
plt.xlabel("Profit")
plt.title("Barplot")
```

A box plot is a method for graphically depicting groups of numerical data through their quartiles. The box extends from the Q1 to Q3 quartile values of the data, with a line at the median (Q2). The whiskers extend from the edges of box to show the range of the data.

```python
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Box Plot")
```

```python
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Box Plot")
```

```python
box1 = np.random.normal(100, 10, 200)   #the same graph in one.
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
plt.show()
plt.boxplot(box2)
plt.show()
```
